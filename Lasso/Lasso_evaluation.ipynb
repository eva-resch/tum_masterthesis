```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Results for Different Regularization Parameters

```python
all_files = os.listdir("Experiments")
row_names = ["p=0.1", "p=0.01", "p=0.001"]
# row_names = ["p=0.1", "p=0.01"]

# --------------------- DAGs ---------------------

# Filter out the files that start with "DLL_DAG_" and end with ".pkl"
dag_files = [f[:-4] for f in all_files if f.startswith("DAG_") and f.endswith(".pkl")]
dag_files_hard = [f for f in dag_files if f.split("_")[2] == "hard"]
dag_files_shift = [f for f in dag_files if f.split("_")[2] == "shift"]
dag_files_no = list(set(dag_files) - set(dag_files_hard) - set(dag_files_shift))
dag_files_no.sort()
dag_files_hard.sort()
dag_files_shift.sort()
DAGs_no = pd.DataFrame(index=row_names, columns=dag_files_no)
DAGs_hard = pd.DataFrame(index=row_names, columns=dag_files_hard)
DAGs_shift = pd.DataFrame(index=row_names, columns=dag_files_shift)

# Iterate over the columns and read the corresponding .pkl file
for index, col in enumerate(dag_files_no):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        DAGs_no.at["p=0.1", col] = df["p=0.1"].mean()
        DAGs_no.at["p=0.01", col] = df["p=0.01"].mean()
        DAGs_no.at["p=0.001", col] = df["p=0.001"].mean()

# Iterate over the columns and read the corresponding .pkl file
for index, col in enumerate(dag_files_hard):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        DAGs_hard.at["p=0.1", col] = df["p=0.1"].mean()
        DAGs_hard.at["p=0.01", col] = df["p=0.01"].mean()
        DAGs_hard.at["p=0.001", col] = df["p=0.001"].mean()

# Iterate over the columns and read the corresponding .pkl file
for index, col in enumerate(dag_files_shift):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        DAGs_shift.at["p=0.1", col] = df["p=0.1"].mean()
        DAGs_shift.at["p=0.01", col] = df["p=0.01"].mean()
        DAGs_shift.at["p=0.001", col] = df["p=0.001"].mean()


# --------------------- cyclic ---------------------

# Filter out the files that start with "cycle_" and end with ".pkl"
cycle_files = [f[:-4] for f in all_files if f.startswith("cycle_") and f.endswith(".pkl")]
cycle_files_hard = [f for f in cycle_files if f.split("_")[1] == "hard"]
cycle_files_shift = [f for f in cycle_files if f.split("_")[1] == "shift"]
cycle_files_no = list(set(cycle_files) - set(cycle_files_hard) - set(cycle_files_shift))
cycle_files_no.sort(key=lambda x: int(x.split('_')[1]) if x != 'cycle_10' else 10)
cycle_files_hard.sort(key=lambda x: int(x.split('_')[2]) if x != 'cycle_hard_10' else 10)
cycle_files_shift.sort(key=lambda x: int(x.split('_')[2]) if x != 'cycle_shift_10' else 10)
cycle_no = pd.DataFrame(index=row_names, columns=cycle_files_no)
cycle_hard = pd.DataFrame(index=row_names, columns=cycle_files_hard)
cycle_shift = pd.DataFrame(index=row_names, columns=cycle_files_shift)

for index, col in enumerate(cycle_files_no):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        cycle_no.at["p=0.1", col] = df["p=0.1"].mean()
        cycle_no.at["p=0.01", col] = df["p=0.01"].mean()
        cycle_no.at["p=0.001", col] = df["p=0.001"].mean()

for index, col in enumerate(cycle_files_hard):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        cycle_hard.at["p=0.1", col] = df["p=0.1"].mean()
        cycle_hard.at["p=0.01", col] = df["p=0.01"].mean()
        cycle_hard.at["p=0.001", col] = df["p=0.001"].mean()

for index, col in enumerate(cycle_files_shift):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        cycle_shift.at["p=0.1", col] = df["p=0.1"].mean()
        cycle_shift.at["p=0.01", col] = df["p=0.01"].mean()
        cycle_shift.at["p=0.001", col] = df["p=0.001"].mean()


# --------------------- non-simple ---------------------

# Filter out the files that start with "nonsimple_" and end with ".pkl"
# Filter out the files that start with "cycle_" and end with ".pkl"
nonsimple_files = [f[:-4] for f in all_files if f.startswith("nonsimple_") and f.endswith(".pkl")]
nonsimple_files_hard = [f for f in nonsimple_files if f.split("_")[1] == "hard"]
nonsimple_files_shift = [f for f in nonsimple_files if f.split("_")[1] == "shift"]
nonsimple_files_no = list(set(nonsimple_files) - set(nonsimple_files_hard) - set(nonsimple_files_shift))
nonsimple_files_no.sort()
nonsimple_files_hard.sort()
nonsimple_files_shift.sort()
nonsimple_no = pd.DataFrame(index=row_names, columns=nonsimple_files_no)
nonsimple_hard = pd.DataFrame(index=row_names, columns=nonsimple_files_hard)
nonsimple_shift = pd.DataFrame(index=row_names, columns=nonsimple_files_shift)

for index, col in enumerate(nonsimple_files_no):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        nonsimple_no.at["p=0.1", col] = df["p=0.1"].mean()
        nonsimple_no.at["p=0.01", col] = df["p=0.01"].mean()
        nonsimple_no.at["p=0.001", col] = df["p=0.001"].mean()

for index, col in enumerate(nonsimple_files_hard):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        nonsimple_hard.at["p=0.1", col] = df["p=0.1"].mean()
        nonsimple_hard.at["p=0.01", col] = df["p=0.01"].mean()
        nonsimple_hard.at["p=0.001", col] = df["p=0.001"].mean()

for index, col in enumerate(nonsimple_files_shift):
    df = pd.read_pickle(f"Experiments/{col}.pkl")
    if all(p in df.columns for p in row_names):
        nonsimple_shift.at["p=0.1", col] = df["p=0.1"].mean()
        nonsimple_shift.at["p=0.01", col] = df["p=0.01"].mean()
        nonsimple_shift.at["p=0.001", col] = df["p=0.001"].mean()
```
